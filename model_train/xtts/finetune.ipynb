```python
fixed_metadata_path = "audio_dataset/metadata.csv"

with open("audio_dataset/metadata.txt", "r", encoding="utf-8") as infile, open(fixed_metadata_path, "w", encoding="utf-8") as outfile:
    for line in infile:
        cols = line.strip().split("|")
        if len(cols) == 2:  # If missing normalized transcript, duplicate the second column
            outfile.write(f"{cols[0]}|{cols[1]}|{cols[1]}\n")
        else:
            outfile.write(line)  # Keep existing lines unchanged

print(f"‚úÖ Fixed metadata file saved as: {fixed_metadata_path}")

```

```python
import os
import librosa

dataset_path = "/home/ahmet/my_projects/realtime_translator_V2_updated/model_train/audio_dataset/wavs"

for file in os.listdir(dataset_path):
    if file.endswith(".wav"):
        wav_path = os.path.join(dataset_path, file)
        try:
            y, sr = librosa.load(wav_path, sr=None)
            duration = librosa.get_duration(y=y, sr=sr)
            print(f"{file} - Duration: {duration:.2f} sec - Sample Rate: {sr} Hz")
        except Exception as e:
            print(f"‚ùå Error loading {file}: {e}")
```
